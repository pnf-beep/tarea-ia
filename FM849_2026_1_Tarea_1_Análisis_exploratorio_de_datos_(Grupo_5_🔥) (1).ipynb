```python

import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

diamonds_filtrado = diamonds[
    ((diamonds["cut"] == "Premium") | (diamonds["cut"] == "Ideal")) &
    (diamonds["x"] > 0) &
    (diamonds["y"] > 0) &
    (diamonds["z"] > 0)
]

diamonds_filtrado["balance"] = diamonds_filtrado["carat"] / diamonds_filtrado["price"]
mejor_diamante = diamonds_filtrado.sort_values(
    by="balance",
    ascending=False
).head(1)

plt.figure()
plt.scatter(diamonds_filtrado["price"], diamonds_filtrado["carat"])
plt.scatter(
    mejor_diamante["price"],
    mejor_diamante["carat"],
    s=100
)

plt.xlabel("Precio del dianmante")
plt.ylabel("Tama√±o en quillates")
plt.title("Relacion entre precio y tama√±o de cada diamante")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset
diamonds = sns.load_dataset("diamonds")

# Filtrar diamantes
diamonds_filtrado = diamonds[
    ((diamonds["cut"] == "Premium") | (diamonds["cut"] == "Ideal")) &
    (diamonds["x"] > 0) &
    (diamonds["y"] > 0) &
    (diamonds["z"] > 0)
].copy()  # <- importante

# Crear variable balance
diamonds_filtrado["balance"] = diamonds_filtrado["carat"] / diamonds_filtrado["price"]

# Encontrar el mejor diamante
mejor_diamante = diamonds_filtrado.sort_values(
    by="balance",
    ascending=False
).head(1)

# Gr√°fico
plt.figure(figsize=(8, 6))
plt.scatter(diamonds_filtrado["price"], diamonds_filtrado["carat"], alpha=0.3)
plt.scatter(
    mejor_diamante["price"],
    mejor_diamante["carat"],
    s=150,
    label="Mejor diamante"
)

plt.xlabel("precio del diamante")
plt.ylabel("tama√±o en quilates")
plt.title("Relaci√≥n entre precio y tama√±o del diamante")
plt.legend()
plt.show()
```
